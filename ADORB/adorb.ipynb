```python
class Component():
    def __init__(self):
        self.CompCat = None
        self.Item = None
        self.Component = None
        self.EstimatedQuantity = None
        self.Unit = None
        self.EstimatedUnitCost = None
        self.EstimatedTotalCost = None
        self.YearLastDone = None
        self.FrequencyOrNormalServiceLife = None
        self.EstimatedRemainingUsefulLife = None
        self.NextYear = None
        self.ComponentAssessmentInCurrentDollars = None

Components = [] #empty array
DataStartsOnLine=22
f=open("Foxhall Components.txt", "r")

LineNum=0
for Line in f:
    LineNum += 1
    if LineNum >= DataStartsOnLine: #and LineNum <= DataStartsOnLine+3:
        # print(Line)
        tmp = Line.split(" ",1)
        Item = tmp[0]
        Rest = tmp[1]
        # print(Item)
        # print(Rest)
        if (Item.endswith("00")): 
            CompCat = Rest.replace('\n','')
            continue
        n = Rest.find("#")
        tmp = Rest.split("#",1)
        Comp = tmp[0]
        Rest = tmp[1]
        #print(Rest)
        tmp = Rest.split(" ",7)
        Qty = tmp[1]
        Unit = tmp[2]
        UnitCost = tmp[4]
        TotalCost = tmp[6]
        Rest = tmp[7]
        #print(Qty)
        #print(Unit)
        #print(UnitCost)
        #print(TotalCost)
        #print(Rest)
        tmp = Rest.split()
        Last=tmp[0]
        Freq=tmp[1]
        RemLife=tmp[2]
        Next=tmp[3]
        Assessment=tmp[5]
        Cmp = Component()
        Cmp.CompCat = CompCat
        Cmp.Item = Item
        Cmp.Component = Comp
        Cmp.EstimatedQuantity = Qty
        Cmp.Unit = Unit
        Cmp.EstimatedUnitCost = UnitCost
        Cmp.EstimatedTotalCost = TotalCost
        Cmp.YearLastDone = Last
        Cmp.FrequencyOrNormalServiceLife = Freq
        Cmp.EstimatedRemainingUsefulLife = RemLife
        Cmp.NextYear = Next
        Cmp.ComponentAssessmentInCurrentDollars = Assessment
        Components.append(Cmp)
ItemCount=len(Components)
print(ItemCount)
txt = '{},{},"{}","{}",{},"${}","${}",{},{},{},{},"${}"'
for n in range(ItemCount):
    print(txt.format(Components[n].CompCat, Components[n].Item, Components[n].Component, 
    Components[n].EstimatedQuantity, Components[n].Unit, Components[n].EstimatedUnitCost, 
    Components[n].EstimatedTotalCost, Components[n].YearLastDone, 
    Components[n].FrequencyOrNormalServiceLife, Components[n].EstimatedRemainingUsefulLife,
    Components[n].NextYear, Components[n].ComponentAssessmentInCurrentDollars))

```
