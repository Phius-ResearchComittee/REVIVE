```python
import plotly.graph_objects as go
import pandas as pd
```

```python

# 1. Load your data
df = pd.read_csv('C:\\Users\\arjun\\REVIVE\\Research\\CATE_2026\\EnthalpyTest4_ResultsTable.csv')
# 2. FILTER: Keep only rows where Battery Size is <= 45
df_filtered = df[df['Cooling Battery Size [kWh]'] <= 45]
# 2. Define data for readability
x_data = df_filtered['dh']
y_data = df_filtered['IDB']
z_data = df_filtered['Cooling Battery Size [kWh]']
# 3. Create the 3D Surface (Mesh)
fig = go.Figure(data=[go.Mesh3d(
    x=x_data,
    y=y_data,
    z=z_data,
    intensity=z_data,         # Colors the surface based on Z values
    colorscale='Hot',     # Color scheme
    opacity=1,              # Slight transparency to see depth
    colorbar_title='Battery Size [kWh]'
)])

# 4. Update Layout for clear labeling
fig.update_layout(
    title='3D Surface: Battery Size vs DH & IDB',
    scene=dict(
        xaxis_title='dh',
        yaxis_title='IDB',
        zaxis_title='Battery Size [kWh]',
        zaxis=dict(range=[35, 45]) 
    ),
    width=900,
    height=700,
    margin=dict(l=0, r=0, b=0, t=50) # Tight layout
)
fig.show()
```
